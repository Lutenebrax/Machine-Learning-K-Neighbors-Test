# Machine Learning - K Neighbors Test

In this project we will use a dataset that contains information like technical aspects about various cars.

Lets try to determine the values required to obtain an accurate model to predict values from a column using the K-nearest Neighbors algorithm.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score, KFold

#pd.set_option("display.max_columns", None)
%matplotlib inline
```

```python
#The dataset didn't contain the headers, so we had to assign them.
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration", "num_doors", "body_style", "drive_wheels", "engine_location", "wheel_base", "length", "width", "height", "curb_weight", "engine_type", "num_cylinders", "engine_size", "fuel_system", "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]

cars = pd.read_csv("imports-85.data", names=headers)
cars
```

```python
cars = cars.replace("?", np.nan)
cars.isnull().sum()
```

```python
cars
```

```python
cars["normalized_losses"] = cars["normalized_losses"].astype(float) 
cars["normalized_losses"].fillna(cars["normalized_losses"].mean(), inplace=True)
```

The dataset now contain some null values that were added replacing the symbol "?". 

We will fill the "normalized_losses" null values with the mean of the rest of the values, and drop the rest of the rows that contain null values.

```python
cars
```

```python
cars = cars.dropna(axis=0).copy()
```

```python
cars
```

```python
cars.isnull().sum()
```

As we can see, now there are no null values on the dataset.

```python
cars["num_doors"] = cars["num_doors"].str.replace("two", "2").str.replace("four", "4")

cars["num_cylinders"] = cars["num_cylinders"].str.replace("three", "3").str.replace("four", "4").str.replace("five", "5").str.replace("six", "6").str.replace("eight", "8").str.replace("twelve", "12")

#For gas cars = 0, and for diesel cars = 1
cars.loc[cars["fuel_type"] == "gas", "fuel_type"] = 0
cars.loc[cars["fuel_type"] == "diesel", "fuel_type"] = 1

#For std cars = 0, and for turbo cars = 1
cars.loc[cars["aspiration"] == "std", "aspiration"] = 0
cars.loc[cars["aspiration"] == "turbo", "aspiration"] = 1
```

Here we replaced some string values in some columns that can be used as intergers, and also changed to 0 or 1 some columns values that only have 2 options(like "gas" and "diesel").

```python
cars
```

```python
cars_cols = cars[["symboling", "normalized_losses", "fuel_type", "aspiration", "num_doors", "wheel_base", "length", "width", "height", "curb_weight", "num_cylinders", "engine_size", "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]].copy()
cars_cols
```

Here we created a new dataset with only the values we will convert to numeric.

```python
cols_int = ["symboling", "normalized_losses", "fuel_type", "aspiration", "num_doors", "curb_weight", "num_cylinders", "engine_size", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]
cols_float = ["wheel_base", "length", "width", "height", "bore", "stroke", "compression_ratio"]

cars_cols[cols_int] = cars_cols[cols_int].astype(int)
cars_cols[cols_float] = cars_cols[cols_float].astype(float)
```

And then we change the variable type to interger or float depending on the values of the column.

```python
cars_cols.info()
```

```python
price_col = cars_cols["price"]

cars_normalized = (cars_cols - cars_cols.min())/(cars_cols.max() - cars_cols.min())
cars_normalized["price"] = price_col

cars_normalized
```

Here we just implemented a min-max normalization on all the columns we will use, except the "price" column, that will be our target for predictions.

```python
def knn_train_test(train_col, target_col, df, k):
    #np.random.seed(1)
    kf = KFold(10, shuffle=True, random_state=1)
    knn = KNeighborsRegressor(n_neighbors=k)
    mses = cross_val_score(knn, df[[train_col]], df[target_col], scoring="neg_mean_squared_error", cv=kf)
    return np.mean(np.sqrt(np.absolute(mses)))
```

This is a function that will cross validate the dataset using 10 folds, and returning the root average of the MSE of each group of folds per column and value for "k".

We decided to use this method using KFold and cross_val_score, instead of just dividing the dataset in 2 like a Holdout Validation. Getting 10 folds or divisions to test the predictions, provide us a better insight to build an accurate model.

```python
#We assign all the columns except "price" (target column for predictions) to 
#the train_cols series
train_cols = cars_normalized.columns.drop('price')

k_values = [1, 3, 5, 7, 9]

for col in train_cols:
    print("\n")
    for k_value in k_values:
        rmses = knn_train_test(col, "price", cars_normalized, k_value)
        print("K Value: ", k_value, "  Column: ", col, "  RMSE: ", rmses)
```

```python
#All columns in un single plot.
plt.figure(figsize=(20, 20))
for col in train_cols:
    all_rmses = []
    for k_value in k_values:
        rmses = knn_train_test(col, "price", cars_normalized, k_value)
        all_rmses.append(rmses)
    #print(all_rmses)
    plt.plot(k_values, all_rmses, label=col)
    plt.xlabel("K Value")
    plt.ylabel("RMSE")
    plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 65))

#Each column on an individual plot.
for col, i in zip(train_cols, range(1, 21)):
    all_rmses = []
    for k_value in k_values:
        rmses = knn_train_test(col, "price", cars_normalized, k_value)
        all_rmses.append(rmses)
    plt.subplot(10, 2, i)
    plt.plot(k_values, all_rmses)
    plt.xlabel("K Value")
    plt.ylabel("RMSE")
    plt.title(col)
plt.show()
```

Here we can see the visualization of each column with each value for K for the KNeighborsRegressor function.

```python
rmses_df = pd.DataFrame(columns=["column", "rmse"])

rmses_df["column"] = [col for col in train_cols]
rmses_df["rmse"] = [knn_train_test(col, "price", cars_normalized, 5) for col in train_cols]
rmses_df = rmses_df.sort_values("rmse")
print(rmses_df)
```

```python
def knn_train_test_multi(train_cols, target_col, df, k):
    kf = KFold(10, shuffle=True, random_state=1)
    knn = KNeighborsRegressor(n_neighbors=k)
    mses = cross_val_score(knn, df[train_cols], df[target_col], scoring="neg_mean_squared_error", cv=kf)
    return np.mean(np.sqrt(np.absolute(mses)))
```

```python
features_ammount = [2, 3, 4, 5]

for i in features_ammount:
    best_features = rmses_df["column"][:i]
    best_rmses = knn_train_test_multi(best_features, "price", cars_normalized, 5)
    print("Ammount of top features: ", i, "   RMSE: ", best_rmses)
```

Here we display the RMSE obtained if we use different ammount of columns/features to predict "price". As we can see even if we use 5 features, the difference in RMSE with the top column/feature (engine_size), is not that noticeable. This indicate that even if we use a higher ammount of features to predict a result, it can provide a more accurate answer.

Now that we have used several features to make predictions, lets try to optimize this method trying different values for k_neighbors.

```python
for i in features_ammount:
    print("\n")
    #we will vary k_neighbors from 1 to 25
    for n in range(1, 26):
        best_features = rmses_df["column"][:i]
        best_rmses = knn_train_test_multi(best_features, "price", cars_normalized, n)
        print("Ammount of Features: ", i, " Ammount N Neighbors: ", n, " RMSE: ", best_rmses)
```

We will do the same process but this time creating a dictionary that help us display a plot.

```python
best_results = {}

for i in features_ammount:
    n_neighbors = {}
    #we will vary k_neighbors from 1 to 25
    for n in range(1, 26):
        best_features = rmses_df["column"][:i]
        best_rmses = knn_train_test_multi(best_features, "price", cars_normalized, n)
        n_neighbors[n] = best_rmses
    best_results[i] = n_neighbors

for key, value in best_results.items():
    print("\nAmmount of best features: ", key)
    for key2, value2 in value.items():
        print(key2, ": ", value2)
```

```python
plt.figure(figsize=(12, 12))
for key, value in best_results.items():
    x = list(value.keys())
    y = list(value.values())
    plt.plot(x, y, label="Ammount of features: {}".format(key))
plt.xlabel("N Neighbors K Value")
plt.ylabel("RMSE")
plt.title("Comparison of Each Set of Ammount of Features with Different K Values")
plt.legend()
plt.show()
```

We can see in this line plot that a fewer N Neighbors K Value seems to provide more accurate predictions, and this done while using 4 or 5 features for training set.  But we must also take into consideration that too few K Values could also cause too much Bias Error. So to maintain a balance between the Bias and Variance Errors, we must take a higher ammount of K Values than 2 or 3.

#### This relation with the Bias and Variance Error can be checked obtaining the Standard Deviation of the RMSE Values of each fold provided by the cross_val_score, and looking for a balance with the Average RMSE Values.
